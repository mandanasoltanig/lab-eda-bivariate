```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns

from scipy.stats import chi2_contingency
```

```python
df = pd.read_csv("amz_uk_price_prediction_dataset.csv")
```

```python
df.head()
```

```python
#Part 1 Best-seller trends across product categories
```

```python
#crosstab analysis
category_bestseller_ct = pd.crosstab(
    df['category'],
    df['isBestSeller'])
print(category_bestseller_ct)
```

```python
#proportion of best-sellers in each category
bestseller_proportion = ( df.groupby('category')['isBestSeller'].mean().sort_values(ascending=False))
print(bestseller_proportion)
```

```python
#chi square test
# Chi-square test
chi2, p, dof, expected = chi2_contingency(category_bestseller_ct)
print("Chi-square statistic:", chi2)
print("P-value:", p)
print("Degrees of freedom:", dof)
```

```python
if p < 0.05:
    print("There is a significant relationship between category and best-seller status.")
else:
    print("Best-seller status is independent of product category.")
```

```python
#cramer's v
# Compute Cramér's V
n = category_bestseller_ct.sum().sum()
cramers_v = np.sqrt(chi2 / (n * (min(category_bestseller_ct.shape) - 1)))
print("Cramér's V:", cramers_v)
```

```python
#stacked bar chart
# normalized crosstab for proportions
category_prop = pd.crosstab(
    df['category'],
    df['isBestSeller'],
    normalize='index'
)
# Sort by proportion of True values
category_prop = category_prop.sort_values(by=True, ascending=False)

# Plot stacked bar chart
category_prop.plot(
    kind='bar',
    stacked=True,
    figsize=(12,6)
)
plt.title('Best-Seller Status by Product Category')
plt.xlabel('Category')
plt.ylabel('Proportion')
plt.xticks(rotation=90)
plt.legend(title='isBestSeller')
plt.tight_layout()
plt.show()
```

```python
#Some product categories have a much higher proportion of best-sellers than others.
#The Chi-square test shows that best-seller status depends on product category.
#Cramér’s V indicates the strength of this relationship.
#The stacked bar chart clearly highlights which categories contain the highest share of best-selling products.
```

```python
### Part 2: Exploring Product Prices and Ratings Across Categories and Brands
```

```python
#0. **Preliminary Step: Remove outliers in product prices.**
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

# Dataframe without outliers
df_no_outliers = df[
    (df['price'] >= lower_limit) &
    (df['price'] <= upper_limit)
]
print(df_no_outliers.shape)
```

```python
#1. **Violin Plots**: Price distribution by category
# Top 20 categories based on count
top20_categories = (df_no_outliers['category'].value_counts().head(20).index)

filtered_df = df_no_outliers[df_no_outliers['category'].isin(top20_categories)]

# Violin plot
plt.figure(figsize=(14,6))
sns.violinplot(data=filtered_df, x='category',y='price')

plt.xticks(rotation=90)
plt.title('Price Distribution Across Categories')
plt.show()
```

```python
#- Which product category tends to have the highest median price? 
#Highest Median Price category
median_prices = (df_no_outliers.groupby('category')['price'].median().sort_values(ascending=False))
print(median_prices.head())
```

```python
#2. **Bar Charts**:

#Average price by category

# Top 10 categories by count

top10_categories = (df_no_outliers['category'].value_counts().head(10).index)

avg_price = ( df_no_outliers[df_no_outliers['category'].isin(top10_categories)]
.groupby('category')['price']
.mean()
.sort_values(ascending=False))

# Bar chart
plt.figure(figsize=(12,6))
avg_price.plot(kind='bar')
plt.title('Average Product Price by Category')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.show()
```

```python
#Highest average price category
overall_avg_price = (df_no_outliers.groupby('category')['price'].mean().sort_values(ascending=False))
print(overall_avg_price.head())
```

```python
#3. **Box Plots**:
#Rating by category
# Top 10 categories by count
top10_rating_categories = (df_no_outliers['category'].value_counts().head(10).index)

ratings_df = df_no_outliers[df_no_outliers['category'].isin(top10_rating_categories)]

# Box plot
plt.figure(figsize=(14,6))
sns.boxplot(
    data=ratings_df,
    x='category',
    y='stars')
plt.xticks(rotation=45)
plt.title('Product Ratings by Category')
plt.show()
```

```python
#Highest median rating category
median_ratings = (df_no_outliers.groupby('category')['stars'].median().sort_values(ascending=False))
print(median_ratings.head())
```

```python
#### Part 3: Investigating the Interplay Between Product Prices and Ratings
#1. **Correlation Coefficients**:
correlation = df_no_outliers['price'].corr(df_no_outliers['stars'])
print("Correlation between price and stars:", correlation)
```

```python
#A negative correlation means higher-priced products tend to receive lower ratings.
```

```python
#scatter plot, price vs rating
plt.figure(figsize=(8,6))
sns.scatterplot(data=df_no_outliers,x='stars', y='price', alpha=0.5)
plt.title('Relationship Between Product Ratings and Prices')
plt.show()
```

```python
#Most products cluster around high ratings.
#Prices vary widely even among similarly rated products.
#There may not be a strong visible relationship between ratings and price.
```

```python
#correlation Heatmap
numerical_df = df_no_outliers.select_dtypes(include=['int64', 'float64'])
corr_matrix = numerical_df.corr()

# Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
#The heatmap shows the strength and direction of relationships between numerical variables.
#Values closer to 1 or -1 indicate stronger correlations.
```

```python
#QQ plot- price distribution
from scipy import stats
plt.figure(figsize=(6,6))
stats.probplot(df_no_outliers['price'], dist="norm", plot=plt)
plt.title('QQ Plot of Product Prices')
plt.show()
```

```python
#Product prices are heavily skewed and not normally distributed.
#Most products are low to moderately priced, with a smaller number of expensive products creating a 
#long tail.
```
